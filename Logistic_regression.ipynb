```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
data = pd.read_csv('/content/bank-full.csv',sep=';')
```

```python
data = data.dropna()

print(data.shape)
print(list(data.columns))
```

```python
data.month = pd.to_datetime(data.month, format = '%b').dt.month
```

```python
data['default'] = data['default'].map({'yes':1, 'no':0})
data['housing'] = data['housing'].map({'yes':1, 'no':0})
data['loan'] = data['loan'].map({'yes':1, 'no':0})
data['y'] = data['y'].map({'yes':1, 'no':0})
```

```python
data
```

```python
data.drop(['poutcome','contact'],axis = 1, inplace = True)
```

```python
data1 =pd.get_dummies(data[['job', 'marital', 'education']])
data.drop(['job', 'marital', 'education'],inplace=True, axis=1)
bank=pd.concat([data, data1], axis=1)
bank
```

```python
bank.columns
```

```python
bank.corr()
```

# data split

```python
from sklearn.model_selection import train_test_split
```

```python
y = bank.y
y
```

```python
x = bank.drop(['y'], axis=1)
x
```

```python
x_train, x_test,y_train,y_test = train_test_split(x , y, random_state = 40, test_size= 0.20)
```

```python
classifier = LogisticRegression()
classifier.fit(x_train,y_train)
```

```python
y_pred = classifier.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```
